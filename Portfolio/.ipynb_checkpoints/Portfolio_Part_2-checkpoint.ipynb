### Data Conversion Of Objects

```python
movies.dtypes
```

```python
convert_data = movies.select_dtypes(include=['object']).copy()
convert_data.head()
```

```python
convert_data["occupation"].value_counts()
```

```python
gender_num = {"gender": {"M": 1, "F": 2 }}

genre_num = {"genre": {"Drama": 1, "Comedy": 2, "Action": 3, "Romance": 4, "Thriller": 5, "Sci-Fi": 6, "Horror": 7, 
                       "Adventure": 8, "Crime": 9, "Children": 10, "Mystery": 11, "War": 12, "Musical": 13, 
                       "Animation": 14, "Western": 15, "Film-Noir": 16, "Documentary": 17,"Fantasy": 18,}}

occupation_num = {"occupation": {"student": 1, "other": 2, "educator": 3, "engineer": 4, "programmer": 5, "administrator": 6, 
                            "writer": 7, "librarian": 8, "technician": 9, "executive": 10, "healthcare": 11, "artist": 12, 
                            "entertainment": 13, "scientist": 14, "marketing": 15, "retired": 16, "lawyer": 17, "salesman": 18, 
                            "doctor": 19, "homemaker": 20,}}
```

```python
convert_data = convert_data.replace(gender_num)
convert_data = convert_data.replace(genre_num)
convert_data = convert_data.replace(occupation_num)
convert_data.head()
```

```python
convert_data.dtypes
```

### Combining Dataframes & Correlations

```python
movies.head()
```

```python
convert_data.head()
```

```python
new_data = movies.select_dtypes(include=['int64', 'float64']).copy()
new_data.head()
```

```python
new_movie_data = pd.concat([new_data, convert_data], axis=1)
new_movie_data.head()
```

### Split Training and Testing Data
* Machine learning models are trained to help make predictions for the future. Normally, we need to randomly split the dataset into training and testing sets, where we use the training set to train the model, and then leverage the well-trained model to make predictions on the testing set. 
* To further investigate whether the size of the training/testing data affects the model performance, please random split the data into training and testing sets with different sizes:
    * Case 1: training data containing 10% of the entire data;
    * Case 2: training data containing 90% of the entire data. 
* Print the shape of training and testing sets in the two cases. 

### Training and Testing Data - 10%

```python
train, test = train_test_split(new_movie_data, test_size=0.1, random_state=142)
```

```python
X_train = train[['userId', 'age', 'movieId', 'timestamp', 'gender', 'occupation', 'genre']]
y_train = train['rating']
X_test = test[['userId', 'age', 'movieId', 'timestamp', 'gender', 'occupation', 'genre']]
y_test = test['rating']
```

```python
# your code and solution
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)
train_pred = reg.predict(X_train)
test_pred = reg.predict(X_test)
```

```python
pred_train = reg.predict(X_train)
mse = ((np.array(y_train)-train_pred)**2).sum()/len(y_train)
r2 = r2_score(y_train, train_pred)
print("Train set performance:")
print("MSE:", mse)
print("R Squared:", r2)


pred_test = reg.predict(X_test)
mse = ((np.array(y_test)-test_pred)**2).sum()/len(y_test)
r2 = r2_score(y_test, test_pred)
print("\nTest set performance:")
print("MSE:", mse)
print("R Squared:", r2)
```

### Train Linear Regression Models with Feature Selection under Cases 1 & 2
* When training a machine learning model for prediction, we may need to select the most important/correlated input features for more accurate results. 
* To investigate whether feature selection affects the model performance, please select two most correlated features and two least correlated features regarding rating, respectively. 
* Train four linear regression models by following the conditions:
    - (model-a) using the training/testing data in case 1 with two most correlated input features
    - (model-b) using the training/testing data in case 1 with two least correlated input features
    - (model-c) using the training/testing data in case 2 with two most correlated input features
    - (model-d) using the training/testing data in case 2 with two least correlated input features
* By doing this, we can verify the impacts of the size of traing/testing data on the model performance via comparing model-a and model-c (or model-b and model-d); meanwhile the impacts of feature selection can be validated via comparing model-a and model-b (or model-c and model-d).    

### Model-A: MovieID Ratings

```python
# your code and solution
X_train = train[['movieId']]
y_train = train['rating']
X_test = test[['movieId']]
y_test = test['rating']
```

```python
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)
train_pred = reg.predict(X_train)
test_pred = reg.predict(X_test)
mean_sq_err = metrics.mean_squared_error(y_train, train_pred)
rmse = np.sqrt(mean_sq_err)
mean_sq_err_2 = metrics.mean_squared_error(y_test, test_pred)
rmse_2 = np.sqrt(mean_sq_err_2)
```

```python
pred_train = reg.predict(X_train)
mse = ((np.array(y_train)-train_pred)**2).sum()/len(y_train)
r2 = r2_score(y_train, train_pred)
print("Train set performance:")
print("MSE:", mse)
print("RMSE:", mse)
print("R Squared:", r2)


pred_test = reg.predict(X_test)
mse = ((np.array(y_test)-test_pred)**2).sum()/len(y_test)
r2 = r2_score(y_test, test_pred)
print("\nTest set performance:")
print("MSE:", mse)
print("RMSE:", rmse_2)
print("R Squared:", r2)
```
